### DATA PREPARATION

```python
# 1. Import the two dataframes 
```

```python
import pandas as pd
data1=pd.read_csv("C:/Users/owner1/Desktop/data/2008-2012.csv")
pd.set_option('display.max_column', 200)
data1.head()
```

```python
import pandas as pd
data2=pd.read_csv("C:/Users/owner1/Desktop/data/psam_p40.csv")
pd.set_option('display.max_column', 200)
data2.head()
```

```python
# 2. Rename the selected columns
```

```python
data1.rename(columns={"SERIALNO":"year","SCHG":"enrollment","SCHL":"Educational attainment","ESR":"Employment Status","SCH":"College Type","SEX":"gender","WAGP":"wage/salary","AGEP":"age"},inplace=True)
data1.head()
```

```python
data2.rename(columns={"SERIALNO":"year","SCHG":"enrollment","SCHL":"Educational attainment","ESR":"Employment Status","SCH":"College Type", "SEX":"gender","WAGP":"wage/salary","AGEP":"age"},inplace=True)
data2.head()
```

```python
# 3. Join the two dataframes
```

```python
a = data1[['year','enrollment','Educational attainment','Employment Status','College Type','gender','wage/salary','age']]
```

```python
b = data2[['year','enrollment','Educational attainment','Employment Status','College Type','gender','wage/salary','age' ]]
```

```python
c=pd.concat([a,b])
```

```python
c.head()
```

```python
c.tail()
```

```python
# 4. Prepare year column
```

```python
c["year"]=(c["year"].round(-9))/1000000000
```

```python
c["year"]=c["year"].astype(int)
```

```python
c.head()
```

```python
c.tail()
```

```python
# 5. Prepare enrollment column
```

```python
c["enrollment"].fillna(0,inplace=True)
c.head()
```

```python
def change_col(x):
    for i in [x]:
        if i ==15:
            return 'undergraduate'
        elif i==16:
            return 'graduate'
        elif i==0:
            return 'not enrolled'
        else:
            return 'enrolled but not in college'
```

```python
c["enrollment"]=c['enrollment'].apply(change_col)
```

```python
c.head()
```

```python
c.tail()
```

```python
replace = {'undergraduate': '1',
           'graduate': '2',
           'not enrolled': '3',
           'enrolled but not in college': '4'}
c['em']=c['enrollment'].replace(replace)
```

```python
c['em']=c['em'].astype('int')
```

```python
c.head()
```

```python
c.tail()
```

```python
c.pivot_table(index ='year', columns ='enrollment',values ='em', aggfunc='count', dropna =True, fill_value=True, margins=True)
```

```python
cpivot = c.pivot_table(index ='year', columns ='enrollment',values ='em', aggfunc='count', dropna =True, fill_value=True, margins=True)
```

```python
cpivot.iloc[:,0:]*100
```

```python
# 6. Prepare Educational attainment column
```

```python
def change_column(x):
    for i in [x]:
        if i ==16:
            return 'high sch/GED'
        elif i==17:
            return 'high sch/GED'
        elif i==18:
            return 'high sch/GED'
        elif i==19:
            return 'high sch/GED'
        elif i==20:
            return 'college degree'
        elif i==21:
            return 'college degree'
        elif i==22:
            return 'college degree'
        elif i==23:
            return 'college degree'
        elif i==24:
            return 'college degree'
        elif i==0:
            return 'no formal education'
        elif i==1:
            return 'no formal education'
        else:
            return 'no diploma'
```

```python
c["Educational attainment"]=c['Educational attainment'].apply(change_column)
c.head()
```

```python
c.tail()
```

```python
replace = {'high sch/GED': '1',
           'college degree': '2',
           'no formal education': '3',
           'no diploma': '4'}
c['ae']=c['Educational attainment'].replace(replace)
```

```python
c.head()
```

```python
c.tail()
```

```python
c.pivot_table(index ='year', columns ='Educational attainment',values ='ae', aggfunc='count', dropna =True, fill_value=True, margins=True)
```

```python
eapivot=c.pivot_table(index ='year', columns ='Educational attainment',values ='ae', aggfunc='count', dropna =True, fill_value=True, margins=True)
```

```python
eapivot.iloc[:,0:]*100
```

```python
# 7. Prepare Employment Status column
```

```python
def change(x):
    for i in [x]:
        if i ==0:
            return 'less than 16 years old'
        elif i==3:
            return 'unemployed'
        elif i==6:
            return 'not in the labor force'
        else:
            return 'employed'
```

```python
c["Employment Status"]=c['Employment Status'].apply(change)
```

```python
c.head()
```

```python
c.tail()
```

```python
replace = {'less than 16 years old': '1',
           'unemployed': '2',
           'not in the labor force': '3',
           'employed': '4'}
c['es']=c['Employment Status'].replace(replace).astype(int)
```

```python
c.head()
```

```python
c.tail()
```

```python
c.pivot_table(index ='year', columns ='Employment Status',values ='es', aggfunc='count', dropna =True, fill_value=True, margins=True)
```

```python
espivot = c.pivot_table(index ='year', columns ='Employment Status',values ='es', aggfunc='count', dropna =True, fill_value=True, margins=True)
```

```python
espivot.iloc[:,0:]*100
```

```python
# 8. Prepare College Type column
```

```python
c["College Type"].fillna(0,inplace=True)
```

```python
c["College Type"]=c["College Type"].astype(int)
```

```python
def col_type(x):
    for i in [x]:
        if i ==3:
            return 'private college'
        elif i==2:
            return 'public college'
        elif i==0:
            return 'less than 3 years old'
        else:
            return 'not attended college in 3 months'
```

```python
c["CT"]=c["College Type"].apply(col_type)
```

```python
c.head()
```

```python
c.tail()
```

```python
c.pivot_table(index ='year', columns ='CT',values ='College Type', aggfunc='count', dropna =True, fill_value=True, margins=True)
```

```python
ctpivot=c.pivot_table(index ='year', columns ='CT',values ='College Type', aggfunc='count', dropna =True, fill_value=True, margins=True)
```

```python
ctpivot.iloc[:,0:]*100
```

```python
# 9. The other variables are ready for use. We just have to make sure they are all intergers
```

```python
c["gender"]=c["gender"].astype(int) #  1 = male and 2 = female
```

```python
def sex(x):
    for i in [x]:
        if i==1:
            return 'male'
        else:
            return 'female'
```

```python
c["sex"]=c["gender"].apply(sex)
```

```python
c
```

```python
c["wage/salary"].fillna(0,inplace=True)
```

```python
c["wage/salary"]=c["wage/salary"].astype(int)
```

```python
c["age"].fillna(0,inplace=True)
```

```python
c["age"]=c["age"].astype(int)
```

```python
c.head()
```

```python
c.tail()
```
